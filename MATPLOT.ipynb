**A line plot is a graphical display of data along a number line with Xs or dots recorded above the responses to indicate the number of occurrences a response appears in the data set**

```python
#### <<<<<<<<<<<< lineplot >>>>>>>>>>>>>>

random_ls=[5,4,8,11,10]
plt.plot(random_ls,marker='*',ms=8,mfc='red') #marker face colour  ,marker size
```

```python
axis_1=[5,4,8,11,10]
axis_2=[6,8,10,15,18]
plt.plot(axis_2,axis_1,marker='*',ms=25,mfc='green')
plt.title('lineplot')
plt.xlabel('axis_1')
plt.ylabel('axis_2')
plt.show
```

```python
axis_1=[5,4,8,11,10]
axis_2=[6,8,10,15,18]
plt.plot(axis_2,axis_1,marker='*',ms=25,mfc='green',color='k')                                 #k is black
plt.title('lineplot')
plt.xlabel('axis_1')
plt.ylabel('axis_2')
plt.show
```

```python
axis_1=[5,4,8,11,10]
axis_2=[6,8,10,15,18]
plt.plot(axis_2,axis_1,marker='*',ms=25,mfc='green',lw=6)                   #line width changing
plt.title('lineplot')
plt.xlabel('axis_1')
plt.ylabel('axis_2')
plt.show
```

```python
axis_1=[5,4,8,11,10]
axis_2=[6,8,10,15,18]
plt.plot(axis_2,axis_1,marker='D',ms=6,mfc='green')               # O,D,p
plt.title('lineplot')
plt.xlabel('axis_1')
plt.ylabel('axis_2')
plt.show
```

**BAR GRAPH :- A bar plot is a graphical representation of data where rectangular bars or columns are used to represent different categories. The height of each bar corresponds to the value it represents.**

```python
country=['japan','america','china','pakistan','india']
population=[25,40,70,10,85]
plt.bar(country,population)
plt.xlabel("COUNTRY")
plt.ylabel("POPULATION")
```

```python
country=['japan','america','china','pakistan','india']
population=[25,40,70,10,85]
plt.bar(country,population)
plt.xlabel("COUNTRY")
plt.ylabel("POPULATION")
plt.bar(country,population,color='green',ec='black')                           #ec edge
```

```python
df=pd.read_csv("/content/Used_Bikes (4).csv")
df
```

```python
brand=list(df["brand"].value_counts().keys())
brand
```

```python
no_of_count=list(df["brand"].value_counts().values)
no_of_count
```

**VERTICAL BAR PLOTTING**

```python
plt.figure(figsize=(10,5))
plt.bar(brand,no_of_count,color='dodgerblue',ec='k',ls='dashed')   #ls --> line style
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict={'fontsize':15})
plt.ylabel('No of bike available',fontdict={'fontsize':15})
plt.title("Brands in showroom",fontdict={'fontsize':15,'color':'Blue'})
# plt.yticks
plt.show()
```

**HORIZANTAL BAR PLOTTING**

```python
plt.figure(figsize=(5,10))
plt.barh(brand,no_of_count,color='dodgerblue',ec='k',ls='dashed')   #ls --> line style
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict={'fontsize':15})
plt.ylabel('No of bike available',fontdict={'fontsize':15})
plt.title("Brands in showroom",fontdict={'fontsize':15,'color':'Blue'})
# plt.yticks
plt.show()
```

**HERE WE CAN CREATE BAR PLOT BY GIVING EDGE SIZE,LENGHT WIDTH,MARKER SIZE,FONTSIZE TO MAKE ATTRACTIVE COLOUR**

```python
plt.figure(figsize=(10,5))
plt.bar(brand,no_of_count,color='green',ec='k',ls='dashed',lw=1.5,width=0.95)   #ls --> line style
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict={'fontsize':15})
plt.ylabel('No of bike available',fontdict={'fontsize':15})
plt.title("Brands in showroom",fontdict={'fontsize':15,'color':'Blue'})
plt.savefig("bike_brands")
# plt.yticks
plt.show()
```

**SCATTER PLOT**:- ***The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for the values of the y-axis . A scatter plot is a diagram where each value in the data set is represented by a dot.***

```python
record_1 = [10,20,30,40,5,60,20,10,45,65,74,85]
record_2 = [85,75,95,4,56,5,25,14,32,36,42,51]

plt.scatter(record_1,record_2)
plt.show()
```
